```python
#Import section
import tensorflow as tf
```

```python
#Define the image data generator
#Didnt use image augmentation as it was not leading to better results
train_loader = tf.keras.preprocessing.image.ImageDataGenerator(
                                                               validation_split=0.25)
```

```python
#Define the train directory. Inside it contains the Same and Different folders mentioned above
train_dir = '/hackathon/train/train_x'
```

```python
#Batch size set to 128. Often 256 worked well too
batch_size = 128
```

```python
#Create the train generator
train_generator = train_loader.flow_from_directory(
                train_dir,  # Source directory for the training images
                batch_size=batch_size,
                target_size=(128, 256),
                class_mode='categorical',
                subset = 'training')
```

```python
#Create the validation generator
validation_generator = train_loader.flow_from_directory(
                train_dir, # Source directory for the validation images
                batch_size=batch_size,
                target_size=(128,256),
                class_mode='categorical',
                subset = 'validation')
```

```python
# Create the base model from the pre-trained model ResNet152
base_model = tf.keras.applications.ResNet152(input_shape=(128,128,3), weights='imagenet', include_top=False)
```

```python
#Freeze the base model
base_model.trainable = False
```

```python
#Since the image pairs are concatenated, divide them to obtain the two images
def divide_data(x):
    x1 = x[:, :, :128, :]
    x2 = x[:, :, 128:, :]
    return x1, x2
```

```python
#The GAP layer
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
```

```python
#Input preprocessing according to the pretrained model used
preprocess_input = tf.keras.applications.resnet.preprocess_input
```

```python
#Model

inputs = tf.keras.Input(shape=(128, 256, 3))
x1, x2 = divide_data(inputs)
#x1 = tf.cast(x1, tf.int32)
#x2 = tf.cast(x2, tf.int32)
x1 = preprocess_input(x1)
x1 = base_model(x1)
x1 = global_average_layer(x1)
x1 = tf.keras.layers.Dropout(0.3)(x1)
x2 = preprocess_input(x2)
x2 = base_model(x2)
x2 = global_average_layer(x2)
x2 = tf.keras.layers.Dropout(0.3)(x2)
x3 = tf.keras.layers.concatenate([x1, x2],
                            axis=-1)
x3 = tf.keras.layers.Dense(1024, activation='relu')(x3)
x3 = tf.keras.layers.BatchNormalization()(x3)
x3 = tf.keras.layers.Dropout(0.3)(x3)
#x3 = tf.keras.layers.Dense(1024, activation='relu')(x3)
#x3 = tf.keras.layers.BatchNormalization()(x3)
#x3 = tf.keras.layers.Dropout(0.25)(x3)
#x3 = tf.keras.layers.Dense(1024, activation='relu')(x3)
#x3 = tf.keras.layers.Dropout(0.25)(x3)
#x3 = tf.keras.layers.BatchNormalization()(x3)
outputs = tf.keras.layers.Dense(2, activation='softmax')(x3)
model = tf.keras.Model(inputs, outputs)
```

```python
#Calculate macro f1 score
import keras.backend as K

def f1(y_true, y_pred):
    def recall(y_true, y_pred):
        """Recall metric.

        Only computes a batch-wise average of recall.

        Computes the recall, a metric for multi-label classification of
        how many relevant items are selected.
        """
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

    def precision(y_true, y_pred):
        """Precision metric.

        Only computes a batch-wise average of precision.

        Computes the precision, a metric for multi-label classification of
        how many selected items are relevant.
        """
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision
    precision = precision(y_true, y_pred)
    recall = recall(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))


```

```python
#Adjust the learning rate
#Compile the model
base_learning_rate = 0.001
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=[f1])
```

```python
#Sets the steps per epoch and the validation_steps
steps_per_epoch = train_generator.n // batch_size
validation_steps = validation_generator.n // batch_size
```

```python
#Save the best model based on validation loss
filepath= "weights-improvement-{epoch:02d}-{val_loss:.2f}.h5"
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True)
callbacks_list = [checkpoint]
```

```python
#In case of 128 batch size, run the model for 50 epochs. For 256, run it for 25 epochs
#To manage the skewness, define the class_weight as required.
epochs = 50
history = model.fit(train_generator,
                              steps_per_epoch = steps_per_epoch,
                              epochs=epochs,
                              workers = 4,
                              class_weight = {1:20, 0:1},
                              callbacks=callbacks_list,
                              validation_data=validation_generator,
                              validation_steps=validation_steps)
```
